```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random values
random_values = np.random.randn(1000)

# Compute sigmoid values
sigmoid_values = 1 / (1 + np.exp(-random_values))

# Compute tanh values
tanh_values = np.tanh(random_values)

# Plot the values
plt.figure(figsize=(14, 7))

# Plot sigmoid values
plt.subplot(1, 2, 1)
plt.plot(random_values, sigmoid_values, 'o', markersize=2)
plt.title('Sigmoid Function')
plt.xlabel('Random Values')
plt.ylabel('Sigmoid Values')

# Plot tanh values
plt.subplot(1, 2, 2)
plt.plot(random_values, tanh_values, 'o', markersize=2, color='orange')
plt.title('Tanh Function')
plt.xlabel('Random Values')
plt.ylabel('Tanh Values')

plt.tight_layout()
plt.show()
```
