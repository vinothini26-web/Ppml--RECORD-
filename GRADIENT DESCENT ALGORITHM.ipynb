```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x, y):
    return x**2 + y**2 + 4

# Define the gradient of the function
def gradient(x, y):
    df_dx = 2 * x
    df_dy = 2 * y
    return np.array([df_dx, df_dy])

# Gradient descent algorithm
def gradient_descent(learning_rate, num_iterations):
    # Start at some random point (x0, y0)
    x, y = np.random.randn(2)
    history = [(x, y, f(x, y))]  # To store the history of (x, y, f(x, y))

    for _ in range(num_iterations):
        grad = gradient(x, y)
        x = x - learning_rate * grad[0]
        y = y - learning_rate * grad[1]
        history.append((x, y, f(x, y)))
    
    return x, y, history

# Parameters
learning_rate = 0.1
num_iterations = 100

# Run gradient descent
x_min, y_min, history = gradient_descent(learning_rate, num_iterations)

# Output the results
print(f"Minimum found at x = {x_min:.4f}, y = {y_min:.4f}")
print(f"Minimum value of f(x, y) = {f(x_min, y_min):.4f}")

# Plotting the history of the optimization
history = np.array(history)

plt.figure(figsize=(14, 6))

# Plot the function value over iterations
plt.subplot(1, 2, 1)
plt.plot(history[:, 2], 'o-')
plt.xlabel('Iteration')
plt.ylabel('f(x, y)')
plt.title('Function Value over Iterations')

# Plot the path taken in the (x, y) plane
plt.subplot(1, 2, 2)
plt.plot(history[:, 0], history[:, 1], 'o-', markersize=4)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Path in (x, y) Plane')

plt.tight_layout()
plt.show()
```
